```python
import pandas as pd
```

```python
# Importe, para um dataframe Pandas, o conteúdo do arquivo CSV

jogadores = pd.read_csv('/content/jogadores.csv', encoding = 'utf-8', delimiter = ',')
```

```python
# Mostre o conteúdo do dataframe criado

jogadores
```

```python
# Remova do dataframe os jogadores que possuem o salário nulo

jogadores = jogadores.dropna(subset = ['salario'])
jogadores
```

```python
# Crie, no dataframe, uma nova coluna, salario_anual, contendo o salário anual de todos
# os jogadores

salario_anual = jogadores['salario']*12
salario_anual

jogadores['salario_anual'] = salario_anual
```

```python
jogadores
```

```python
# Atualize o nome dos jogadores, passando todos os caracteres para maiúsculo

jogadores['jogador'] = jogadores['jogador'].str.upper()
```

```python
jogadores
```

```python
# Mostre os dados estatísticos dos valores numéricos do dataframe (comando describe).
# Descreva as informações sobre os salários dos jogadores, obtidas através desses dados.

print(jogadores.describe())
```

```python
# Faça consultas no dataframe para exibir:
# - Nome e time dos jogadores que possuem salário acima de R$ 200000,00

jogadores.query('salario > 200000')[['jogador', 'time']]
```

```python
# - Nome e salário dos jogadores dos times de Minas Gerais (estado MG)

jogadores.query('estado == \'MG\'')[['jogador', 'salario']]
```

```python
# - Nome e time dos jogadores cujo nome contenha a letra 'u'

jogadores[jogadores['jogador'].str.lower().str.contains('u')]
```

```python
# - Nome, salário e time dos jogadores, ordenados pelo salário em ordem descrescente

jogadores.sort_values(by = 'salario', ascending = False)[['jogador', 'salario', 'time']]
```

```python
# - Nome, salário e time dos jogadores, ordenados pelo nome do time em ordem crescente,
# e em seguida pelo salário em ordem decrescente

ordenadoPorTime = jogadores.sort_values(by = 'time')
ordenadoPorTime.sort_values(by = 'salario', ascending = False)[['jogador', 'salario', 'time']]
```

```python
# - A quantidade de jogadores por time

jogadores['time'].value_counts()
```

```python
# - A média salarial por time

jogadores.groupby(['time']).mean(numeric_only=True)
```

```python
# Exporte o conteúdo do dataframe para um novo arquivo CSV, nomeado novo_time.csv

jogadores.to_csv('novo_time.csv', encoding = 'utf-8', index = True)
```
